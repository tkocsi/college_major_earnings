```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
```

```python
recent_grads.describe()
```

```python
raw_data_count = recent_grads.shape[0]
print(raw_data_count)
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
print(cleaned_data_count)
```

```python
recent_grads.plot(x='Total', y='Median', kind='scatter', title='Total vs. Median', xlim=(-100,4500))
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
```

```python
recent_grads['Sample_size'].hist(bins=15, range=(0,5000))
```

```python
cols=['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']
fig = plt.figure(figsize=(10,25))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r-1]].hist(xrot=40)
    ax.set_xlabel(cols[r-1])
    
```

```python
cols=['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']
fig = plt.figure(figsize=(10,40))
for r in range(4, len(cols)-2):
    ax = fig.add_subplot(len(cols)-2,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40)
    ax.set_xlabel(cols[r])
```

```python
ax = recent_grads['Women'].hist(figsize=(10,5), bins=10, range=(0,350000), xrot=40)
ax.set_xlabel('Women')
```

```python
ax=recent_grads['Men'].hist(figsize=(10, 5), bins=10, range=(0,350000), xrot=40)
ax.set_xlabel('Men')
```
